# Validation without pydantic

```python
# student class
# name, age, gender

class Student:
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender
    def __repr__(self):
        return (f"Sudent(name= '{self.name}', age= {self.age}, gender= '{self.gender}')")

s1 = Student("Alex", 40, "M")

s1
```
