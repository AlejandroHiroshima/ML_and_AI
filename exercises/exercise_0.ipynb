In the data folder, there are a few job ads taken from arbetsf√∂rmedlingen.se.

a) Read all the jobs ads into python

b) Create a function that uses gemini to summarize a job ad. This function should take in an ad as its parameter and return a summary.

c) Now create and export markdown files for each job ad and its corresponding summary.

d) Try to take other job ads from arbetsf√∂rmedlingen.se and see how well your function performs.

```python
# a) Read all the jobs ads into Python

ads = ["ads1", "ads2", "ads3"]
read_ads=[]
for ad in ads:
    with open(f"data/{ad}.txt", "r", encoding="utf-8") as file:
        ad_print = file.read()
        read_ads.append(ad_print)


print(read_ads[0])
```

```python
""" b) Create a function that uses gemini to summarize a job ad. 
 This function should take in an ad as its parameter and return a summary."""

from google import genai
from google.genai import types
client = genai.Client()
system_instruction = """
    You are a helpful assistant, you summarize whatever I ask you to summarize. Keep it short.
"""

def summarize_ad(ad, model= "gemini-2.5-flash"):
    response = client.models.generate_content(
        model=model,
        contents=ad,
        config= types.GenerateContentConfig(
        system_instruction=system_instruction
    ))
    return response.text

summarized_ads = [summarize_ad(ad) for ad in read_ads]

print(summarized_ads[0])
    
```
